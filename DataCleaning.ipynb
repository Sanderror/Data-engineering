```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data/Phishing_Legitimate_full.csv")
data
```

```python
data.describe()
```

```python
# Create correlation heatmap
# Create correlation matrix
corr_matrix = data.corr()

# Create figure
plt.figure(figsize=(30, 24))

# Create heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')

# Show plot
plt.show()
```

```python
# Select relevant columns for ML application
# Selection is based on the interpretability of the variables, as they should be easily identifiable by an individual
selected_columns = ['CLASS_LABEL', 'NumDots', 'UrlLength', 'NumDash', 'NumDashInHostname', 'AtSymbol', 'TildeSymbol', 'NumUnderscore', 'NumPercent', 'NumAmpersand', 'NumHash', 'NumNumericChars', 'NoHttps']

# Create new dataframe with only the relevant variables
data_selected  = data.loc[:, selected_columns]
data_selected
```

```python
data_selected.describe()
```

```python
# Create correlation heatmap
# Create correlation matrix
corr_matrix = data_selected.corr()

# Create figure
plt.figure(figsize=(30, 24))

# Create heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')

# Show plot
plt.show()
```
